```python
import pandas as pd
import os
import random
```

```python
# 导入数据
df_query = pd.DataFrame(columns = ['日期','排名','热搜词','热度'])
for filename in sorted(os.listdir('../爬虫/分小时数据')):
    path = '../爬虫/分小时数据/' + filename
    new = pd.read_excel(path)
    df_query = pd.concat([df_query,new],ignore_index=True)

df_query['date'] = df_query['日期'].apply(lambda x:x[:13])
df_query['month'] = df_query['日期'].apply(lambda x:x[:7])
df_query.head()
```

# 分层抽样

```python
sample = pd.DataFrame(columns = ['日期','排名','热搜词','热度'])
month = '2020-06'
df = df_query[df_query['month']==month]
df['day'] = df['date'].apply(lambda x:x[:10])
df.head()
```

```python
dates = []
while len(dates) < 7:
    date = random.choice(list(df['date'].unique()))
    if float(date[-2:]) >9 and float(date[-2:]) <21:
        continue
    if date in dates:
        continue
    dates.append(date)
dates.sort()
for date in dates:
    new = df[df['date']==date][['日期','排名','热搜词','热度']]
    sample = pd.concat([sample,new],ignore_index=True)
    
sample.to_excel("分层抽样-202006.xlsx",index=False)
sample.head()
```

```python
sample = pd.DataFrame(columns = ['日期','排名','热搜词','热度'])
month = '2021-07'
df = df_query[df_query['month']==month]
dates = random.sample(list(df['date'].unique()),7)
dates.sort()
for date in dates:
    new = df[df['date']==date][['日期','排名','热搜词','热度']]
    sample = pd.concat([sample,new],ignore_index=True)
    
sample.to_excel("分层抽样-202107.xlsx",index=False)
sample.head()
```

```python
sample = pd.DataFrame(columns = ['日期','排名','热搜词','热度'])
months = list(df_query['month'].unique())
for month in months:
    df = df_query[df_query['month']==month]
    dates = random.sample(list(df['date'].unique()),7)
    dates.sort()
    for date in dates:
        new = df[df['date']==date][['日期','排名','热搜词','热度']]
        sample = pd.concat([sample,new],ignore_index=True)
```

```python
sample.to_excel("分层抽样.xlsx",index=False)
sample.head()
```

# 系统抽样

```python
sample = pd.DataFrame(columns = ['日期','排名','热搜词','热度'])
dates = list(df_query['date'].unique())
N  = len(dates)
for i in range(1,N,97):
    date = dates[i]
    new = df_query[df_query['date']==date][['日期','排名','热搜词','热度']]
    sample = pd.concat([sample,new],ignore_index=True)
```
